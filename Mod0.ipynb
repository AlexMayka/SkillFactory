```python
def guess_the_number(report_info):
    '''Function for finding a number (upgraded algorithm)
        Алгоритм половинного деления (берется сред. знач. в промежутке и меняется в зависимости от задуманного числа)
        для поиска числа (улучшенный, по сравнению с примером)
    '''
    
    from random import randint
    
    print("Введите желаемое кол-во испытаний")
    test=int(input())
    
    count_ls = [] #list for the average number of attempts (для поиска сред. значения кол-ва попыток)
    
    for i in range(test):
    
        left_point = 0 #the leftmost point (левая точка интервала)
        right_point = 100 #the rightmost point (правая точка интервала)
                
        hidden_num = randint(left_point, right_point) #random number on the interval (случайное число на интервале)
        attempt = 0 #number of attempts in this test (количество попыток в этом тесте)
                
        while True:
            attempt += 1
            midd_num = round( (left_point+right_point) / 2 ) #middle value in the interval (среднее значение на интервале)
                    
            if (midd_num > hidden_num): 
                right_point = midd_num #changes the right interval (меняем левую часть интервала)
                
            elif (midd_num < hidden_num):
                left_point = midd_num #changes the left interval (меняем правую часть интервала)
                
            else:
                
                if attempt in report_info.keys(): #add the attempt to the dictionary(заносим кол-во попыток в словарь для отчета)
                    report_info[attempt] += 1  
                else:
                    report_info[attempt] = 1
                    
                count_ls.append(attempt) #add the attempt to the list(заносим кол-во попыток в лист)
                break
            
            
    sred_attemp = (round (sum(count_ls) / len(count_ls)) ) #count the average value of attempts and round it up (вычисляем среднее кол-во попыток)
    print ("За", test ,"испытаний среднее кол-во попыток =", sred_attemp,'\n\n')
    return report_info
    
    
    
def report(report_info):
    '''Function for viewing statistics
        Функция для показа статистики ( за сколько кол-во попыток угадано сколько испытаний)
    '''
    for key,num in report_info.items():
        print ("За", key ,"попыток угадано",num,"испытаний\n\n")
    return 1
    
    

def menu():
    '''Start the program using the menu
        Запуск программы через меню
    '''
    
    command = 0 #The command number (номер комманды)
    check = False #Checking for a report (проверка для функции report
    report_info = {} #information for the report (информация для отчета)
    
    
    print("Добро пожаловать!!!")
    print("Введите номер комманду(начните с №1):\n\n")
    
    while True: #while not 3.Выход
        print("Меню:")
        print("1.Начать тестирование")
        print("2.Показать отчет")
        print("3.Выход\n\n")
        print("Введите цифру: ", end=" ")
        
        command=int(input()) 
        
        if (command == 1):
            guess_the_number(report_info)
            check=True
            
        elif ((command == 2) and (check == True)):
            report(report_info)
            
        elif (command == 3):
            print("Всего доброго))")
            return 1
           
        else:
            print("Странно,не могу выполнить. Введите другую команду...")

    
    
    
menu()
```
